```python
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
sb.set()

print('Problem 1')
traindata = pd.read_csv('train.csv')
print("Data type: ",type(traindata))
print('')
print('1(b)')
print(traindata.dtypes)
print("")

print('1(c): integer variables data frame')
integerData = traindata.select_dtypes(include=np.number).columns.tolist()
print(integerData)
print("")

print('1(d/e): real integer variables data frame')
realIntData = pd.DataFrame(traindata[['Id', 'LotFrontage', 'LotArea', 'YearBuilt', 'YearRemodAdd', 'MasVnrArea', 'BsmtFinSF1', 'BsmtFinSF2', 'BsmtUnfSF', 'TotalBsmtSF', '1stFlrSF', '2ndFlrSF', 'LowQualFinSF', 'GrLivArea', 'BsmtFullBath', 'BsmtHalfBath', 'FullBath', 'HalfBath', 'BedroomAbvGr', 'KitchenAbvGr', 'TotRmsAbvGrd', 'Fireplaces', 'GarageYrBlt', 'GarageCars', 'GarageArea', 'WoodDeckSF', 'OpenPorchSF', 'EnclosedPorch', '3SsnPorch', 'ScreenPorch', 'PoolArea', 'MiscVal', 'MoSold', 'YrSold', 'SalePrice']])
print(realIntData.columns.tolist)
```

```python
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
sb.set()

print('Problem 2')
print('2a: Summary statistics from Sale Price')
print(traindata['SalePrice']. describe())
saleprice = pd.DataFrame(traindata['SalePrice'])

print('')
print('2b: visualise statitistics')
print('boxplot')
b_plot = (saleprice.boxplot(column = 'SalePrice' ))
b_plot.plot()
print('histogram')
h_plot = (saleprice.hist(column = 'SalePrice'))
print('kdeplot')
kde_plot = saleprice.plot.kde()
print('')

```

```python
print('2c Lot Area summary')
lotarea = pd.DataFrame(traindata["LotArea"])
print(traindata['LotArea']. describe())

print('')
print('2d: visualise statitistics')
print('boxplot')
b_plot = (lotarea.boxplot(column = 'LotArea' ))
b_plot.plot()
print('histogram')
h_plot = (lotarea.hist(column = 'LotArea'))
print('kdeplot')
kde_plot = lotarea.plot.kde()
print('')
```

```python
print('2e sale price vs lot area joint plot')
jointdf = pd.concat([saleprice,lotarea],axis = 1).reindex(saleprice.index)
sb.jointplot(data = jointdf, x = "LotArea", y = "SalePrice", height = 12)
print(jointdf.corr())
```
